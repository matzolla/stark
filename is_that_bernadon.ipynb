```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python

data =pd.read_excel('sinistre.xlsx')
```

```python
data.head()
```

```python
data.tail()
```

```python
# Create an empty list
Row_list =[]
  
# Iterate over each row
for index, rows in data.iterrows():
    # Create list for the current row
    my_list =[rows["en milliards d'euros"], rows[2009],rows[2010],rows[2011],rows[2012],rows[2013],rows[2014],rows[2015],rows[2016],rows[2017],rows[2018],rows[2019],rows[2020]]
      
    # append the list to the final list
    Row_list.append(my_list)
  
# Print the list
print(Row_list)
```

```python
# Initialise the subplot function using number of rows and columns
X=[2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
figure, axis = plt.subplots(3, 3,figsize=(18,10))
  
# I'am plotting 9 plots 
axis[0, 0].plot(X, Row_list[0][1:])
axis[0, 0].set_title(Row_list[0][0])
  
axis[0, 1].plot(X, Row_list[1][1:])
axis[0, 1].set_title(Row_list[1][0])


axis[0, 2].plot(X, Row_list[2][1:])
axis[0, 2].set_title(Row_list[2][0])

axis[1, 0].plot(X, Row_list[3][1:])
axis[1, 0].set_title(Row_list[3][0])

axis[1, 1].plot(X, Row_list[4][1:])
axis[1, 1].set_title(Row_list[4][0])


axis[1, 2].plot(X, Row_list[5][1:])
axis[1, 2].set_title(Row_list[5][0])
  
axis[2, 0].plot(X, Row_list[6][1:])
axis[2, 0].set_title(Row_list[6][0])

axis[2, 1].plot(X, Row_list[7][1:])
axis[2, 1].set_title(Row_list[7][0])

axis[2, 2].plot(X, Row_list[8][1:])
axis[2, 2].set_title(Row_list[8][0])
plt.show()
```
